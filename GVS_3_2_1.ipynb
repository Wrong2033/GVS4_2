```python
import torch
import torch.nn as nn
import torch.nn.init as init
import math

import linearlayer

class MyLinearLayer(torch.autograd.Function):
    @staticmethod
    def forward(ctx, x, w, b):
        # Сохранение входных данных для использования в backward
        ctx.save_for_backward(x, w, b)

        # Вызываем CUDA-функцию для прямого прохода
        y = linearlayer.my_forward_linear(x, w, b)
        return y

    @staticmethod
    def backward(ctx, grad_output):
        # Извлекаем сохраненные данные
        x, w, b = ctx.saved_tensors

        # Вызываем CUDA-функцию для обратного прохода
        grad_input, grad_weight, grad_bias = linearlayer.my_backward_linear(x, w, b, grad_output)
        return grad_input, grad_weight, grad_bias

    @staticmethod
    def reset_parameters(w, b):
        # Инициализация весов и смещений
        init.kaiming_uniform_(w, a=math.sqrt(5))
        fan_in, _ = init._calculate_fan_in_and_fan_out(w)
        bound = 1 / math.sqrt(fan_in) if fan_in > 0 else 0
        init.uniform_(b, -bound, bound)
```

```python
class LinearModel(nn.Module):
    def __init__(self, in_features, out_features):
        super(LinearModel, self).__init__()
        self.in_features = in_features
        self.out_features = out_features
        # Создаем параметры на GPU
        self.w = nn.Parameter(torch.Tensor(out_features, in_features).cuda())
        self.b = nn.Parameter(torch.Tensor(out_features).cuda())

        # Инициализация параметров
        MyLinearLayer.reset_parameters(self.w, self.b)

    def forward(self, x):
        return MyLinearLayer.apply(x, self.w, self.b)
```

```python
class CustomLinear(nn.Module):
    def __init__(self, in_features, out_features, bias=True):
        super(CustomLinear, self).__init__()
        self.in_features = in_features
        self.out_features = out_features
        self.weight = nn.Parameter(torch.empty(out_features, in_features))
        self.bias = nn.Parameter(torch.empty(out_features)) if bias else None
        self.reset_parameters()

    def reset_parameters(self):
        init.kaiming_uniform_(self.weight, a=math.sqrt(5))
        if self.bias is not None:
            fan_in, _ = init._calculate_fan_in_and_fan_out(self.weight)
            bound = 1 / math.sqrt(fan_in) if fan_in > 0 else 0
            init.uniform_(self.bias, -bound, bound)

    def forward(self, input):
        return MyLinearLayer.apply(input, self.weight, self.bias)
```

```python
import torch
import torch.nn as nn
import torch.nn.init as init
import math
import linearlayer


# Пользовательская реализация линейного слоя с использованием CUDA
class MyLinearLayer(torch.autograd.Function):
    @staticmethod
    def forward(ctx, x, w, b):
        """
        Прямой проход пользовательского слоя.
        :param ctx: Контекст для сохранения промежуточных данных для backward.
        :param x: Входные данные (batch_size x in_features).
        :param w: Матрица весов (out_features x in_features).
        :param b: Смещения (out_features).
        :return: Результат линейной операции (batch_size x out_features).
        """
        # Убедимся, что входные тензоры континуальны (требуется для CUDA-операций)
        x = x.contiguous()
        w = w.contiguous()
        b = b.contiguous()

        # Сохраняем тензоры для обратного прохода
        ctx.save_for_backward(x, w, b)

        # Вызываем функцию прямого прохода из CUDA-модуля
        y = linearlayer.my_forward_linear(x, w, b)
        return y

    @staticmethod
    def backward(ctx, grad_output):
        """
        Обратный проход для вычисления градиентов.
        :param ctx: Контекст, содержащий сохраненные входные тензоры.
        :param grad_output: Градиенты на выходе (batch_size x out_features).
        :return: Градиенты для x, w, b.
        """
        # Извлекаем сохраненные тензоры
        x, w, b = ctx.saved_tensors

        # Убедимся, что все тензоры континуальны
        grad_output = grad_output.contiguous()
        x = x.contiguous()
        w = w.contiguous()
        b = b.contiguous()

        # Вызываем функцию обратного прохода из CUDA-модуля
        grad_input, grad_weight, grad_bias = linearlayer.my_backward_linear(x, w, b, grad_output)
        return grad_input, grad_weight, grad_bias

    @staticmethod
    def reset_parameters(w, b):
        """
        Инициализация параметров слоя.
        :param w: Матрица весов.
        :param b: Смещения.
        """
        # Используем стандартные методы PyTorch для инициализации
        init.kaiming_uniform_(w, a=math.sqrt(5))
        fan_in, _ = init._calculate_fan_in_and_fan_out(w)
        bound = 1 / math.sqrt(fan_in) if fan_in > 0 else 0
        init.uniform_(b, -bound, bound)


# Обертка для пользовательского линейного слоя
class LinearModel(nn.Module):
    def __init__(self, in_features, out_features):
        """
        Инициализация линейного слоя.
        :param in_features: Размер входного слоя.
        :param out_features: Размер выходного слоя.
        """
        super(LinearModel, self).__init__()
        self.in_features = in_features
        self.out_features = out_features

        # Параметры слоя создаются как тензоры на GPU
        self.w = nn.Parameter(torch.Tensor(out_features, in_features).cuda())
        self.b = nn.Parameter(torch.Tensor(out_features).cuda())

        # Инициализация параметров
        MyLinearLayer.reset_parameters(self.w, self.b)

    def forward(self, x):
        """
        Прямой проход модели.
        :param x: Входные данные (batch_size x in_features).
        :return: Результат линейной операции.
        """
        return MyLinearLayer.apply(x, self.w, self.b)


# Класс для тестирования пользовательского линейного слоя
class TestLinearLayer:
    def __init__(self):
        """
        Инициализация тестового класса.
        """
        self.tolerance = 1e-5  # Допустимая погрешность при сравнении результатов

    def run_test(self):
        """
        Запуск тестов на совпадение результатов пользовательской реализации с PyTorch.
        """
        # Параметры модели
        in_features = 10  # Размер входных данных
        out_features = 5  # Размер выходных данных
        batch_size = 2  # Размер мини-выборки

        # Генерация тестового входа
        x = torch.randn(batch_size, in_features, device="cuda", requires_grad=True)

        # Стандартная линейная модель PyTorch
        torch_linear = nn.Linear(in_features, out_features).cuda()
        torch_output = torch_linear(x)
        torch_output.sum().backward()  # Прямой и обратный проход

        # Пользовательская модель
        custom_linear = LinearModel(in_features, out_features).cuda()

        # Копируем веса и смещения из PyTorch-модели в пользовательскую модель
        custom_linear.w.data = torch_linear.weight.detach().clone()
        custom_linear.b.data = torch_linear.bias.detach().clone()

        # Прямой и обратный проход для пользовательской модели
        custom_output = custom_linear(x)
        custom_output.sum().backward()

        # Проверка совпадения результатов прямого прохода
        torch.testing.assert_close(
            custom_output, torch_output, atol=self.tolerance, rtol=self.tolerance
        )

        # Проверка совпадения градиентов весов
        torch.testing.assert_close(
            custom_linear.w.grad, torch_linear.weight.grad, atol=self.tolerance, rtol=self.tolerance
        )

        # Проверка совпадения градиентов смещений
        torch.testing.assert_close(
            custom_linear.b.grad, torch_linear.bias.grad, atol=self.tolerance, rtol=self.tolerance
        )

        # Проверка совпадения градиентов входных данных
        torch.testing.assert_close(
            x.grad, x.grad, atol=self.tolerance, rtol=self.tolerance
        )

        print("Тест успешно пройден!")


# Запуск теста
test = TestLinearLayer()
test.run_test()
```
